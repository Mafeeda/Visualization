```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [-3, 5, 7]
```

```python
y = [10, 2, 5]
```

```python
x
```

```python
y
```

```python
fig = plt.figure(figsize = (10,4))
plt.plot(x,y)
```

```python
fig = plt.figure(figsize = (15,3))

plt.plot(x, y)
plt.xlim(0, 10)
plt.ylim(-3, 8)
plt.xlabel('X Axis')
plt.ylabel('Y axis')
plt.title('Line Plot',size = 20)
plt.suptitle('Sales Comaprison', size=15)
```
